```python
inventory = {}

def addProduct(productId,name,price,quantity):
    if productId in inventory:
        print("product already exists")
    else:
        inventory[productId] = {'name':name,'price':price,'quantity':quantity}
        print("product successfully added")
              
def removeProduct(productId):
    if productId in inventory:
        del inventory[productId]
        print("product successfully deleted")
    else:
        print("product does not exists")
        
        
def updateProduct(productId,name,price,quantity):
    if productId in inventory:
        product = inventory[productId]
        if name:
            product['name'] = name
        if price:
            product['price'] = price
        if quantity:
            product['quantity'] = quantity
        print("product updated succesfully")
    else:
        print("product does not exist")


def displayInventory():
    print("Inventory")
    for productId,product in inventory.items():
        print(f"productId :{productId}")
        print(f"product name:{product['name']}")
        print(f"product price:{product['price']}")
        print(f"product quantity:{product['quantity']}")
        print()
        
def checkProductAvailability(productId,product):
    if productId in inventory:
        print("Product exist in inventory")
    else:
        print("Product doesn't exists")
        
        
N = int(input("Enter the number of products:"))

for i in range(N):
    productId = int(input("Enter the productId:"))
    name = input("Enter the name of the product:")
    price = float(input("Enter the price of the product:"))
    quantity = int(input("Enter the quantity of the product:"))
    addProduct(productId,name,price,quantity)
    print()

displayInventory()


removeProduct(int(input("Enter the productId to remove:")))
displayInventory()


print("Update a product in the inventory:")
updateProduct(int(input()),input(),float(input()),int(input()))
displayInventory()
```

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("country_vaccinations (1).csv")
df.head()
```

```python
df = df.dropna()
```

```python
df_jan = df[df['month']=='01']
df_feb = df[df['month']=='02']
df_mar = df[df['month']=='03']
df_apr = df[df['month']=='04']
df_may = df[df['month']=='05']
df_jun = df[df['month']=='06']
df_july = df[df['month']=='07']
df_aug = df[df['month']=='08']
df_sep = df[df['month']=='09']
df_oct = df[df['month']=='10']
df_nov = df[df['month']=='11']
df_dec = df[df['month']=='12']
```

### No.of Vaccinations monthly in 2021

```python
print("January :",df_jan['total_vaccinations'].sum())
print("February :",df_feb['total_vaccinations'].sum())
print("March :",df_mar['total_vaccinations'].sum())
print("April :",df_apr['total_vaccinations'].sum())
print("May :",df_may['total_vaccinations'].sum())
print("June :",df_jun['total_vaccinations'].sum())
print("July :",df_july['total_vaccinations'].sum())
print("Augest :",df_aug['total_vaccinations'].sum())
print("September :",df_sep['total_vaccinations'].sum())
print("October :",df_oct['total_vaccinations'].sum())
print("November :",df_nov['total_vaccinations'].sum())
print("December :",df_dec['total_vaccinations'].sum())
```

```python
months = ['Jan','Feb','Mar','Apr','May','June','July','Aug','Sep','Oct','Nov','Dec']
freq= [df_jan['total_vaccinations'].sum(),df_feb['total_vaccinations'].sum(),
         df_mar['total_vaccinations'].sum(),
         df_apr['total_vaccinations'].sum(),
         df_may['total_vaccinations'].sum(),
         df_jun['total_vaccinations'].sum(),
         df_july['total_vaccinations'].sum(),
         df_aug['total_vaccinations'].sum(),
         df_sep['total_vaccinations'].sum(),
         df_oct['total_vaccinations'].sum(),
         df_nov['total_vaccinations'].sum(),
         df_dec['total_vaccinations'].sum()]

print(months)
print(freq)
```

```python
plt.plot(months,freq,color='b',marker='*',linewidth=0.8,linestyle='dashed',markersize=4)
plt.xlabel('Month')
plt.ylabel("Total vaccinations")
plt.title("Total No.of Vaccinations monthly")
plt.show()
```

```python
plt.bar(months,freq,color='brown',width=0.5)
plt.xlabel('Month')
plt.ylabel("Total vaccinations")
plt.title("Total No.of Vaccinations monthly")
plt.show()
```

```python
explodes = [.2,.1,.4,.01,.01,.01,.01,.01,.01,.01,.01,.01]
plt.pie(freq,labels=months,startangle=0,explode=explodes,shadow=True,autopct="%1.2f%%")
#plt.legend()
plt.show()
```

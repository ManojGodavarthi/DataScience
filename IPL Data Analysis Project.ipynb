### which team has the best scoring run-rate in the first 6 overs?

```python
rr=bb[bb['over']<7].groupby('batting_team').sum().sort_values(ascending=False,by='total_runs')
```

```python
overs = list(bb[bb['over']<7].groupby('batting_team').sum().sort_values(ascending=False,by='total_runs')['over'])
runs = list(bb[bb['over']<7].groupby('batting_team').sum().sort_values(ascending=False,by='total_runs')['total_runs'])
run_rate1=[]
for i in range(len(overs)):
    run_rate1.append(round(overs[i]/runs[i],2))
rr['RunRate']=run_rate1
```
